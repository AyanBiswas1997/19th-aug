```python
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
x=df.iloc[:,3:13]
y=df.iloc[:,13]
```

```python
y
```

```python
x.head()
```

```python
Geography=pd.get_dummies(x['Geography'],drop_first=True)
Gender=pd.get_dummies(x['Gender'],drop_first=True)
```

```python
x.head()
```

```python
x=x.drop(['Geography','Gender'],axis=1)
```

```python
x.head()
```

```python
x=pd.concat([x,Geography,Gender],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
x_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU,PReLU,ELU,LeakyReLU
from tensorflow.keras.layers import Dropout
```

```python
Classifier=Sequential()
```

```python
# adding input layer
Classifier.add(Dense(units=11,activation='relu'))
```

```python
#adding 1st layer
Classifier.add(Dense(units=7,activation='relu'))
```

```python
#adding 2nd layer
Classifier.add(Dense(units=6,activation='relu'))
```

```python
#adding output layer
Classifier.add(Dense(1,activation='sigmoid'))
```

```python
Classifier.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
import tensorflow as tf

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.001,
    patience=20,
    verbose=0,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model_history=model_history=Classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=1000,callbacks=early_stopping)
```

```python
model_history.history.keys()
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.xlabel('eppoch')
plt.ylabel('accuracy')
```

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.xlabel('eppoch')
plt.ylabel('accuracy')
```

```python
##prediction model
y_pred=Classifier.predict(x_test)
y_pred=y_pred>=0.5
```

```python
#confution_matrix and accuracy_score
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)
```
